## Investigation Overview

> The goal of this investigation is to better understand the nature of conflict in Brazil by examining conflict events that took place between Jan 1, 2019 and May 4, 2020. The main focus is fatality counts, locations and the category of the event.

## Dataset Overview

> This dataset was aquired via the [Humanitarian Data Exchange (HDX)](https://data.humdata.org/) and shared by the [Armed Conflict Location & Event Data Project (ACLED)](http://www.acleddata.com/). The dataset contains over 14,000 records, including characteristics about the type, agents, exact location, and date of the event.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
data = pd.read_csv('conflict_data_bra_clean.csv')
```

## Distribution of Fatalities

> In this dataset, a total of 6,609 fatalities happened across 14,222 observations. Fortunately, a high number of events involved no fatalities, but approximately 11 had a fatality count over 8 people. Further investigation reveals that the high fatality events are mostly concentrated in the northern regions of Brazil.

```python
fig, ax = plt.subplots(figsize=(12,2))

sns.boxplot(x=data['fatalities'], ax=ax)
ax.set_title("Fatality Counts Per Event")
ax.set_xlabel('Fatality Count')
ax.tick_params(labelsize=10);
```

## Conflict Locations

#### Rio de Janeiro state leads by a large margin in events and fatalities.

> Rio de Janeiro state not only has the highest number of records in the dataset; it also has the highest count of fatalities. In fact, the top 7 states with the most records in the dataset are also the states with the deadliest conflicts.

```python
fig, ax = plt.subplots(figsize=(14,8))
data.groupby('state')[['fatalities']].sum().plot(kind='barh', 
                                                 ax=ax,
                                                 title='Total Fatalities by State',
                                                 legend=False);
```
